```python
import nltk
from nltk.tokenize import sent_tokenize

# Variable text yang sudah diubah
text = """Python is a powerful programming language. It is widely used in data science and web development. Many developers love its simple syntax."""

print("Original string:")
print(text)

token_text = sent_tokenize(text)

print("\nSentence-tokenized copy in a list:")
print(token_text)

print("\nRead the list:")
for sentence in token_text:
    print(sentence)
```

```python
import nltk
from nltk.tokenize import sent_tokenize

# Teks baru dalam bahasa Jerman
text = '''
Künstliche Intelligenz verändert die Welt schnell. Viele Unternehmen investieren viel Geld in diese neue Technologie. 
Es gibt jedoch auch wichtige ethische Bedenken, die wir diskutieren müssen.
'''

print("\nOriginal string:") 
print(text)

# Tokenisasi kalimat dengan parameter bahasa Jerman
token_text = sent_tokenize(text, language='german') 

print("\nSentence-tokenized copy in a list:") 
print(token_text)

print("\nRead the list:")
for sentence in token_text:
    print(sentence)
```
